# **Probabilidad de los intervalos de confianza**

**$P=\frac{1}{\sigma\sqrt{2\pi}}\int_{\bar{x}-\sigma}^{\bar{x}+\sigma}e^{-\frac{(x-\bar{x})^{2}}{2\sigma^{2}}}dx$**

```python
std_dev = 1
constante=1/(std_dev*(2*np.pi)**(1/2))
mean=0
f = lambda x:constante*np.exp(-((x-mean)**2)/(2*std_dev**2))
quad(f,mean-std_dev,mean+std_dev)
```

```python
x=np.linspace(-10,10,500)
y=f(x)
plt.plot(x,y)
```
